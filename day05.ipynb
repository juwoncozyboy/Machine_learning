```python
data = {
    '이름' : ['짱구', '철수', '유리', '맹구', "짱아"],
    '성별' : ['남', '남', '여', '남', '여'],
    '나이' : [15, 14, None, 10, 5],
    '키' : [150.1, 163.2, 100.2, None, 80.3],
    '몸무게' : [40, 43, 34, 38, None],
    '학생여부' : [True, True, False, None, False]
}

df = pd.DataFrame(data)
df.info()
df
```

```python
#fillna(): 결측값을 특정 값으로 채움
df = df.fillna(0.0)
df
```

```python
# groupby() : 열을 기준으로 데이터를 그룹화함
# 성별열을 기준으로 그룹화하여 키의 평균값을 구함
print(df.groupby(['성별'])[['키']].mean())

# sort_values() : 열의 값을 기준으로 데이터를 정렬함
# 나이열을 기준으로 오름차순 정렬
print(df.sort_values('나이')) #오름차순정렬
print(df.sort_values('나이', ascending = False)) #내림차순정렬

# rename() : 열 또는 행의 이름을 변경함
# 키 열의 이름을 Height로 변경
df.rename(columns={'키':'Height'}, inplace = True)
df
```

```python
data = {
    '이름' : ['짱구', '철수', '유리', '맹구', "짱아","짱아"],
    '성별' : ['남', '남', '여', '남', '여','여'],
    '나이' : [15, 14, None, 10, 5,5],
    '키' : [150.1, 163.2, 100.2, None, 80.3, 80.3],
    '몸무게' : [40, 43, 34, 38, None, None],
    '학생여부' : [True, True, False, None, False, False]
}

df = pd.DataFrame(data)

df
```

```python
# drop_duplicates() : 중복된 행을 제거함
df = df.drop_duplicates()
df

# pivot() : 데이터 프레임의 열을 행 인덱스와 열로 반환함
# 이름 열을 행 인덱스로 변경
#df.pivot(index = '이름', columns = '성별', values = '나이')

# merge() : 두 개 이상의 데이터 프레임을 병합함
df2 = pd.DataFrame({'과목' : ['국어', '영어', '수학', '사회', '과학'],
                   '학점' : [4.5, 3.7, 3.9, 4.2, 4.4],
                    '이름' : ['짱구', '유리', '맹구', '짱아', '철수']})
df2

merge_df = pd.merge(df, df2, on = "이름")
merge_df

# apply() : 함수를 통해 데이터 프레임의 행 또는 열에 작업을 수행
# 키열과 몸무게 열의 평균값을 구하기
# print(merge_df[['키', '몸무게']].apply(np.mean))

# map() : 열의 각 요소에 함수나 매핑 딕셔너리를 적용하여 값을 변환함
# 성별열의 값을 1로 변환함
#df['성별']
#df.info()
df['성별'] = df['성별'].map({'남': 1, '여': 0})
print(df.info())
print(df)
```

```python
data = {
    '이름' : ['짱구', '철수', '유리', '맹구', "짱아", "짱아"],
    '성별' : ['남', np.nan, '여', '남', '여', '여'],
    '나이' : [15, 14, None, 10, 5, 5],
    '키' : [150.1, 163.2, 100.2, None, 80.3, 80.3],
    '몸무게' : [40, 43, 34, 38, None, None],
    '학생여부' : [True, True, False, None, False, False]
}
#np.nan : NaN 값을 나타내는 Numpy 함수

df = pd.DataFrame(data)
#Nan값을 2로 대체
df['성별'] = df['성별'].fillna(2)

#성별열의 값을 남 : 1과 여 : 0으로 변환
df['성별'] = df['성별'].map({2:2, '남':1, '여':0})

df
```

```python
# astype(): 열의 데이터 타입을 변경함
# "나이열의 데이터타입을 문자열로 변경"
df['나이'] = df['나이'].astype(str)
df.info()

#나이 열의 타입 확인
print(df['나이'].dtype) #데이터프레임, 시리즈 데이터 타입 확인

# agg() : 그룹화된 데이터에 대해 집계함수를 적용(sum, mean, max 등)
#print(df.groupby('이름'))[['나이','키','몸무게']].agg()
#fillna()-> 0.0변경 후 나이, 키, 몸무게의 평균값 구하기
df.fillna(0.0, inplace = True)

#이름을 기준으로 나이, 키 , 몸무게의 평균 구하기
avg = df.groupby('이름')[['키', '몸무게']].agg(np.mean)
avg
```
