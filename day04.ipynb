```python
data = {'이름':['짱구', '철수', '유리','훈이', '맹구'],
        '학년':[1, 2, 1, 3, 2],
        '점수':[80, 70, 90, 85, 78]}
df = pd.DataFrame(data)
df
```

```python
#1. 점수가 가장 높은 학생의 이름 출력하기
max_score = df['점수'].max()
st = df.loc[df['점수'] == max_score, '이름'].values[0]
print(st)

#.idxmax() 최대값의 인덱스를 반환하는 메소드
st1 = df.loc[df['점수'].idxmax(), '이름']
print(st1)

#2. 점수 열 내림차순 정렬하기
score_df = df.sort_values(by='점수', ascending=False)
print(score_df)

#3. 학년 열을 기준으로 각 평균 점수 출력하기(.groupby()나 .unique()를 사용해서 가능함)
grade = df['학년'].unique() #학년 열의 고유한값들을 리스트
print(grade, type(grade))

for i in grade:
    avg = df[df['학년'] == i]['점수'].mean()
    print(f"{i}학년의 평균점수 : {avg}")

#방법2) groupby이용
avg = df.groupby('학년')['점수'].mean()
print(avg)

for i, j in avg.items(): #.items() 학년별로 구한 평균점수를 딕셔너리 반환
    print(f"{i}학년 평균점수 : {j}")
#4. 학년이 2이상인 학생들 정보 출력하기
print(df[df['학년']>=2])
#5. 전체 평균점수 출력하기
total = df['점수'].mean
print(total)
#df_score = df['점수']
#mean_score = df_score.mean()
#mean_score
```

##8)pie()

```python
figure = plt.figure()
axes = figure.add_subplot(111)

labels = ['봄', '여름', '가을', '겨울']
ratio = [50, 30, 12, 28] #100을 기준으로 각 값이 몇퍼센트를 차지하는지 출력

axes.pie(ratio, labels = labels, autopct='%.3f%%',
         wedgeprops = {'linewidth':2, 'edgecolor':'black'})
#autopct는 부채 꼴안에 표시될 숫자의 형태 지정
#%.2f 소수점 2자리까지 출력, %% -> %를 표시
#wedgerprops 옵션안에서 딕셔너리 형태로 지정
#linewidth 테두리의 두께, edgecolort 테두리 색상 지정
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

labels = ['봄', '여름', '가을', '겨울']
ratio = [50, 30, 12, 28] #100을 기준으로 각 값이 몇퍼센트를 차지하는지 출력

#여름 조각 빼내기
explode = [0, 0.2, 0, 0]
#값은 0이상의 실수

axes.pie(ratio, labels = labels, autopct = '%.1f%%', explode = explode)

plt.show()
```

```python
#데이터프레임 생성
data_df = pd.DataFrame({'항목':['a', 'b', 'c'], '값':[30, 50, 20]})
# data_df
# data_df['항목'] #

#2행 2열 서브플롯
figure, axes = plt.subplots(2, 2)

# figure = plt.figure()
# axes1 = figure.add_subplot(2, , 1)

#첫번째 서브플롯(2행 2열의 첫 번째)
axes[0, 0].plot([1, 2, 3], [4, 5, 6])
axes[0, 0].set_title('서브플롯1')

#두번째 서브플롯(2행 2열의 두번째)
axes[0, 1].scatter([1, 2, 3], [4, 5, 6])
axes[0, 1].set_title('서브플롯2')

#세번째 서브플롯(2행 2열의 세번째)
axes[1, 0].bar(data_df['항목'], data_df['값'])
axes[1, 0].set_title('서브플롯3')

#네번째 서브플롯(2행 2열의 네번째)
axes[1, 1].pie(data_df['값'], labels = data_df['항목'], autopct = '%.2f%%')
axes[1, 1].set_title('서브플롯4')

figure.suptitle('2행 2열의 서브플롯', fontsize = 15)

plt.tight_layout()
plt.show()
```
