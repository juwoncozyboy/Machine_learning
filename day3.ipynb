##2 시리즈(Series)

```python
import pandas as pd
di_data = {'서울':1, '경기':2, '강원':3}
li_data = ['20230723', 723, 'python', 100, True]

#시리즈로 변경
s1 = pd.Series(di_data)
s2 = pd.Series(li_data)

print(s1)
print(s2)
```

```python
print(s1)

#추가
s1['부산'] = 4
print(s1)
#s1.append(pd.Series(['제주'])) append 메소드로 ['값'] 값이 추가

#수정
s1['서울'] = 10
print(s1)

s1 = s1.drop('강원')
print(s1)
```

```python
print(s2)

# 인덱스 5에 해당하는 위치에 값을 추가
s2[5] = 100
print(s2)

# 인덱스 2에 해당하는 값을 pandas로 수정
s2[2] = 'Pandas'
print(s2)
s2.at[0] = 20230724
print(s2)

#5번째 값 삭제
s2 = s2.drop(5)
print(s2)
```
